```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from conway import conway, conway_iter
```

```python
# create board and "Pulsar" seed

board = np.zeros((17,17))

#top left
board[7,4:7] = 1 #bottom line
board[4:7,7] = 1 #right line
board[4:7,2] = 1 #left line
board[2,4:7] = 1 #top line

#bottom left
board[9,4:7] = 1 #top line
board[10:13,7] = 1 #right line
board[10:13,2] = 1 #left line
board[14,4:7] = 1 #bottom line

#top right
board[4:7,9] = 1 #left line
board[7,10:13] = 1 #bottom line
board[4:7,14] = 1 #right line
board[2,10:13] = 1 #top line

#bottom right
board[9,10:13] = 1 #top line
board[10:13,9] = 1 #left line
board[10:13,14] = 1 #right line
board[14,10:13] = 1 #bottom line

plt.imshow(board)
plt.show()
```

```python
conway_iter(board,N_iter=3,saveimgs=False)
```

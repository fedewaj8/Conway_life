```python
# create board and "Pulsar" seed

board = np.zeros((17,17))

#top left
board[7,4:7] = 1 #bottom line
board[4:7,7] = 1 #right line
board[4:7,2] = 1 #left line
board[2,4:7] = 1 #top line

#bottom left
board[9,4:7] = 1 #top line
board[10:13,7] = 1 #right line
board[10:13,2] = 1 #left line
board[14,4:7] = 1 #bottom line

#top right
board[4:7,9] = 1 #left line
board[7,10:13] = 1 #bottom line
board[4:7,14] = 1 #right line
board[2,10:13] = 1 #top line

#bottom right
board[9,10:13] = 1 #top line
board[10:13,9] = 1 #left line
board[10:13,14] = 1 #right line
board[14,10:13] = 1 #bottom line

plt.imshow(board)
plt.show()
```

```python
plt.imshow(board, interpolation='nearest')
plt.show()
```

```python
# iterate pulsar
for i in range(4):
    plt.imshow(board,interpolation='nearest')
    plt.pause(0.01)
    board = conway(board)
    
plt.show()
```

```python
np.random.seed(2)
board = np.random.randint(0,2,(50,50))
plt.imshow(board)
plt.show()
```

```python
def conway_iter(board,N_iter,t_pause=0.01,saveimgs=False):
    '''
    '''
    i=0
    for i in range(N_iter):
        plt.figure(figsize=(10,10))
        plt.imshow(board,interpolation='nearest')
        
        if saveimgs == True:
            plt.savefig('conway_rand_{:0>3}'.format(i))
            i+=1
        plt.pause(t_pause)
        board = conway(board)
        
    plt.show()
    
```

```python
conway_iter(board,N_iter=3,saveimgs=True)
```
